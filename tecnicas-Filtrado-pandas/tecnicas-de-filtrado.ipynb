```python
#funcion que suma los puntos de un piloto
def suma_puntos(df,driver):
    piloto = df[df['Driver'] == driver]
    suma_puntos = piloto['PTS'].sum()
    print(f'{piloto} sumo la cantidad de: {suma_puntos} puntos')
   
```

```python
suma_puntos(df,'Max Verstappen VER')
```

```python
#funcion que sume todos los puntos de un pais en todos los años
def suma_Puntos_Nacionalidad(df,nacionalidad):
    nacion = df[df['Nationality'] == nacionalidad]
    puntos = nacion['PTS'].sum()
    return puntos
```

```python
suma_Puntos_Nacionalidad(df,'ITA')
```
